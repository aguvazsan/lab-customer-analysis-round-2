```python
# Reemplazamos los valores nulos en la columnas 
df_2['state'] = df_2['state'].fillna('Unknow')
df_2['response'] = df_2['response'].fillna('Unknow')
df_2['vehicle class'] = df_2['vehicle class'].fillna('Unknow')
df_2['vehicle size'] = df_2['vehicle size'].fillna('Unknow')
df_2['vehicle type'] = df_2['vehicle type'].fillna('Unknow')
```

```python
round(df_2.isna().sum()/len(df_2),4)*100
```

```python
# Para comenzar la limpieza visualizamos si existen filas que contengan mas de un valor nulo

df_2.isna().sum(axis=1)
```

```python
# Reemplazo los valores nulos de las columnas 'months since last claim' 'number of open complaints' y por su media

df_2['months since last claim']= df_2['months since last claim'].fillna(df_2['months since last claim'].mean())
df_2['number of open complaints']= df_2['number of open complaints'].fillna(df_2['number of open complaints'].mean())
```

```python
round(df_2.isna().sum()/len(df_2),4)*100
```

```python
#Convertimos la columna 'effective to date' en formato fecha

df_2['effective to date'] = pd.to_datetime(df_2['effective to date'])
```

```python
#Creamos una columna con el mes

df_2['Month'] = df_2['effective to date'].dt.month

df_2.head
```

```python
# Filtramos la información unicamente por el primer trimestre

q1_data = df_2.loc[(df_2['Month'] >= 1) & (df_2['Month'] <= 3)]
```
